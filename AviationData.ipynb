###   Data Visualization

```python
#Distribution of Injury across different aircraft categories 

plt.figure(figsize=(12, 6))
sns.boxplot(x='Aircraft.Category', y='Total.Fatal.Injuries', data=df_relevant)
plt.xticks(rotation=45)
plt.title('Distribution of Fatal Injuries by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Total Fatal Injuries')
plt.show()
```

```python
# Distribution Category
aircraft_category_distribution = df['Aircraft.Category'].value_counts()
print(aircraft_category_distribution)


plt.figure(figsize=(10, 6))
sns.barplot(x=aircraft_category_distribution.index, y=aircraft_category_distribution.values)
plt.title('Distribution of Aircraft Categories')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Incidents')
plt.show()
```

```python
#Distribution of weather Conditions 
weather_condition_distribution = df['Weather.Condition'].value_counts()
print(weather_condition_distribution)

plt.figure(figsize=(8, 8))
plt.pie(weather_condition_distribution, labels=weather_condition_distribution.index, autopct='%1.1f%%', 
        colors=sns.color_palette('Set2'), wedgeprops=dict(width=0.4, edgecolor='w'))
plt.title('Proportion of Incidents by Weather Condition')
plt.show()
```

```python
#Distribution of incidents by Broad Phase of Flight
phase_of_flight_distribution = df['Broad.phase.of.flight'].value_counts()
print(phase_of_flight_distribution)


plt.figure(figsize=(10, 6))
sns.barplot(x=phase_of_flight_distribution.index, y=phase_of_flight_distribution.values)
plt.title('Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)
plt.show()
```

```python
injury_columns = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']
df[injury_columns].describe()

# injury severity using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[injury_columns])
plt.title('Distribution of Injuries in Incidents')
plt.xlabel('Injury Type')
plt.ylabel('Number of Injuries')
plt.show()

# distribution of Aircraft Damage
aircraft_damage_distribution = df['Aircraft.damage'].value_counts()
print(aircraft_damage_distribution)

# aircraft damage distribution using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=aircraft_damage_distribution.index, y=aircraft_damage_distribution.values)
plt.title('Distribution of Aircraft Damage')
plt.xlabel('Damage Type')
plt.ylabel('Number of Incidents')
plt.show()
```
